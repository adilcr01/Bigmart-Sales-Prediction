# Preprocessing Task before Model Building

## 1) Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df_train['item_fat_content']= le.fit_transform(df_train['item_fat_content'])
df_train['item_type']= le.fit_transform(df_train['item_type'])
df_train['outlet_size']= le.fit_transform(df_train['outlet_size'])
df_train['outlet_location_type']= le.fit_transform(df_train['outlet_location_type'])
df_train['outlet_type']= le.fit_transform(df_train['outlet_type'])
```

```python
df_train
```

```python
df_train
```

# 2) Splitting our data into train and test

```python
X=df_train.drop('item_outlet_sales',axis=1)
```

```python
Y=df_train['item_outlet_sales']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, random_state=101, test_size=0.2)
```

# Model Building

### Linear Regression

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
```

```python
lr.fit(X_train_std,Y_train)
```

```python
X_test.head()
```

```python
Y_pred_lr=lr.predict(X_test_std)
```

```python
print(r2_score(Y_test,Y_pred_lr))
print(mean_absolute_error(Y_test,Y_pred_lr))
print(np.sqrt(mean_squared_error(Y_test,Y_pred_lr)))
```

```python
joblib.dump(lr,r'lr.sav')
```

### RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
rf= RandomForestRegressor(n_estimators=1000)
```

```python
rf.fit(X_train_std,Y_train)
```

```python
Y_pred_rf= rf.predict(X_test_std)
```

```python
print("R2_Score :",r2_score(Y_test,Y_pred_rf))
print("MAE :",mean_absolute_error(Y_test,Y_pred_rf))
print("MSE :",np.sqrt(mean_squared_error(Y_test,Y_pred_rf)))
```

```python
joblib.dump(lr,r'rf.sav')
```

### AdaBoostRegressor

```python
from sklearn.ensemble import AdaBoostRegressor

model_adb= AdaBoostRegressor(n_estimators = 100)
model_adb.fit(X_train_std,Y_train)

# predicting the test set results
Y_pred_adb = model_adb.predict(X_test_std)

# RMSE
mse = mean_squared_error(Y_test, Y_pred_adb)
print("R2_Score :",r2_score(Y_test,Y_pred_adb))
print("MAE :",mean_absolute_error(Y_test,Y_pred_adb))
print("RMSE :", np.sqrt(mse))
```

```python
joblib.dump(model_adb,r'adb.sav')
```

### GradientBoostingRegressor

```python
from sklearn.ensemble import GradientBoostingRegressor

model_gbr = GradientBoostingRegressor()
model_gbr.fit(X_train_std,Y_train)

# predicting the test set results
y_pred_gbr = model_gbr.predict(X_test_std)
#print(y_pred_gbr)

# RMSE
mse = mean_squared_error(Y_test, y_pred_gbr)

# Calculating the root mean squared error
print("R2_Score :",r2_score(Y_test,y_pred_gbr))
print("MAE :",mean_absolute_error(Y_test,y_pred_gbr))
print("RMSE :", np.sqrt(mse))
```

```python
joblib.dump(model_gbr,r'gbr.sav')
```

### DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor

model_dtr = DecisionTreeRegressor()
model_dtr.fit(X_train_std,Y_train)

# predicting the test set results
y_pred_dtr = model_dtr.predict(X_test_std)
#print(y_pred)

# RMSE
mse = mean_squared_error(Y_test, y_pred_dtr)

# Calculating the root mean squared error
print("R2_Score :",r2_score(Y_test,y_pred_dtr))
print("MAE :",mean_absolute_error(Y_test,y_pred_dtr))
print("RMSE :", np.sqrt(mse))
```

```python
joblib.dump(model_dtr,r'dtr.sav')
```

### Support vector machine


```python
from sklearn.svm import SVR

model_svr = SVR()
model_svr.fit(X_train_std,Y_train)

# predicting the x test results
y_pred_svr = model_svr.predict(X_test_std)

# RMSE
mse = mean_squared_error(Y_test, y_pred_svr)

# Calculating the root mean squared error
print("R2_Score :",r2_score(Y_test,y_pred_svr))
print("MAE :",mean_absolute_error(Y_test,y_pred_svr))
print("RMSE :", np.sqrt(mse))
```

```python
joblib.dump(model_svr,r'svr.sav')
```

### XgBoost Regressor

```python
from xgboost import XGBRegressor

model_xgb = XGBRegressor()
model_xgb.fit(X_train_std,Y_train)

# predicting the test set results
y_pred_xgb = model_xgb.predict(X_test_std)
#print(y_pred)

# RMSE
mse = mean_squared_error(Y_test, y_pred_xgb)

# Calculating the root mean squared error
print("R2_Score :",r2_score(Y_test,y_pred_xgb))
print("MAE :",mean_absolute_error(Y_test,y_pred_xgb))
print("RMSE :", np.sqrt(mse))
```

```python
joblib.dump(model_xgb,r'xgb.sav')
```

# Hyper-Parameter Optimization

```python
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from sklearn import metrics
```

```python
#Randomized Search CV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
# Create the random grid

#XG Boost
params_XGB={ 
    "n_estimators"     :[10,50,100,150,250,350,500],
    "learning_rate"    :[0.001, 0.05, 0.1, 0.5, 0.75, 0.1],
    "max_depth"        :[   1,    2,   3],
    "min_child_weight" :[   1,    3,    5,    7],
    "gamma"            :[0.0,   0.1,  0.2,  0.3, 0.4],
    "colsample_bytree" :[0.3,   0.4,  0.5,  0.7]
}


# Gradient Boost
params_GBR={
    'loss' :['ls','lad', 'huber','quantile'],
    'n_estimators' :[100, 150, 200, 250,350,500],
    'learning_rate'  :[0.01, 0.05, 0.1, 0.15, 0.2],
    'max_depth'  :[1,2,3],
    'min_samples_split':[2,3,4,5],
    'min_samples_leaf' :[1, 2, 3, 5, 7, 10],
}


# Random Forest
random_rf= {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
```

# Randomized Search CV

### Random Forest Hyper Parameter Optimization

```python
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_rf,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = -1)
```

```python
rf_random.fit(X_train_std,Y_train)
```

```python
rf_random.best_params_
```

```python
prediction_rf = rf_random.predict(X_test_std)
```

```python
plt.figure(figsize = (8,8))
sns.distplot(Y_test-prediction_rf)
plt.show()
```

```python
plt.figure(figsize = (8,8))
plt.scatter(Y_test, prediction_rf, alpha = 0.5)
plt.xlabel("y_test")
plt.ylabel("y_pred")
plt.show()
```

```python
print("R2_Score :",r2_score(Y_test,prediction_rf))
print('MAE:', metrics.mean_absolute_error(Y_test, prediction_rf))
print('MSE:', metrics.mean_squared_error(Y_test, prediction_rf))
print('RMSE:', np.sqrt(metrics.mean_squared_error(Y_test, prediction_rf)))
```
